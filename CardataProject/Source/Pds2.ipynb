```python
import pandas as pd
```

```python
data=pd.read_csv("/content/sample_data/train.csv")
```

```python
data.head()
```

```python
#Examine any null values.
```

```python
data.isna().sum()
```

```python
data2 = pd.DataFrame(data)
```

```python
# (b) Mileage, Engine, Power, and New Price units were removed and converted to floats.
```

```python
data2['Mileage'] = data2['Mileage'].str.extract('(\d+\.\d+)').astype(float)

data2['Engine'] = data2['Engine'].str.replace(' CC', '').astype(float)

data2['Power'] = data2['Power'].str.extract('(\d+\.\d+)').astype(float)

data2['New_Price'] = data2['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

```python
# (a)
#Substitute the NA values for the engine, power, and mileage with the median
#The best substitute method is the median since it prevents outliers and maintains the center tendency.
#Change the Seat NA settings to Mode. Despite being a quantitative value, the number of seats is discontinuous, hence the mode is employed.
#It is preferable to remove a column in the instance of Newprice since there are a lot of null values there. Inputing any values could generate outliers in the data and incorrect analysis.
```

```python
data2['Mileage'].fillna(data2['Mileage'].median(),inplace=True)
data2['Engine'].fillna(data2['Engine'].median(),inplace=True)
data2['Power'].fillna(data2['Power'].median(),inplace=True)

data2['Seats'].fillna(data2['Seats'].mode()[0],inplace=True)

data2.drop('New_Price', axis=1, inplace=True)#axis=1 indicates to drop a column
```

```python
output_csv_file = 'cl_data.csv'
data2.to_csv(output_csv_file, index=False)
```
