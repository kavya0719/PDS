```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/d4.csv")
```

```python
# Histograms for math, reading, and writing scores
plt.figure(figsize=(12, 4))
plt.subplot(131)
plt.hist(data['math score'], bins=20, edgecolor='k')
plt.title('Math Score Histogram')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(132)
plt.hist(data['reading score'], bins=20, edgecolor='k')
plt.title('Reading Score Histogram')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.subplot(133)
plt.hist(data['writing score'], bins=20, edgecolor='k')
plt.title('Writing Score Histogram')
plt.xlabel('Score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Bar plot for race/ethnicity distribution
race_counts = data['race/ethnicity'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=race_counts.index, y=race_counts.values, palette='viridis')
plt.title('Race/Ethnicity Distribution')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Count')
plt.show()
```

```python
# Violin plots to compare scores by gender
plt.figure(figsize=(10, 5))
plt.subplot(131)
sns.violinplot(data=data, x='gender', y='math score', palette='Set2')
plt.title('Math Score by Gender')

plt.subplot(132)
sns.violinplot(data=data, x='gender', y='reading score', palette='Set2')
plt.title('Reading Score by Gender')

plt.subplot(133)
sns.violinplot(data=data, x='gender', y='writing score', palette='Set2')
plt.title('Writing Score by Gender')

plt.tight_layout()
plt.show()
```

```python
# count plot for test preparation course
plt.figure(figsize=(6, 4))
sns.countplot(data=data, x='test preparation course', palette='Set3')
plt.title('Test Preparation Course Completion')
plt.xlabel('Test Preparation Course')
plt.ylabel('Count')
plt.show()
```

```python
#  pairplot with hue for race/ethnicity
sns.pairplot(data=data, vars=['math score', 'reading score', 'writing score'], hue='race/ethnicity', palette='Dark2')
plt.suptitle('Pairplot of Scores by Race/Ethnicity', y=1.02)
plt.show()
```
